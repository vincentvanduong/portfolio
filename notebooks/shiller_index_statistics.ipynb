```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Specify variable used for Plotting
file_path = 'C:/Users/vince/Documents/GitHub/automation/data/shiller.csv'
```

```python
df = pd.read_csv(file_path)
df = df.iloc[::-1]
df
```

```python
# Defining Plots
def plot_timeseries(data):

    ax1 = plt.subplot2grid((5, 1), (0, 0), rowspan=4, colspan=1)
    plt.title('Shiller PE Ratio')
    plt.grid()

    ax1.plot(data, color='r')


    # remove timestamps
    ax1.axes.get_xaxis().set_visible(False)
    #ax1.legend([symbol1 + " Low Price", symbol1 + " High Price", symbol2 + " Low Price", symbol2 + " High Price"], loc='upper right')


def volume_plot(data, symbol1, symbol2):
    '''
    This function is very similar to above's function with the exception
    of creating a smaller subplot and using different data. This function
    is meant for displaying volumes. 
    '''
    ax2 = plt.subplot2grid((5, 1), (4, 0), rowspan=1, colspan=1)
    plt.title('Bid & Ask Volumes Development of the Stocks ' + symbol1 + " and " + symbol2)
    plt.grid()
    ax2.plot(data.index, data[('Volume',symbol1)], color='r')
    ax2.plot(data.index, data[('Volume',symbol2)], color='g')

    # remove timestamps
    ax2.axes.get_xaxis().set_visible(False)
    ax2.legend([symbol1 + ' Volume', symbol2 + ' Volume'], loc='upper right')
```
